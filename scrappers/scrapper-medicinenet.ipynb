```python
import pandas as pd
import numpy as np
import requests # library to handle requests
from bs4 import BeautifulSoup # library to parse HTML documents
```

```python
wikiurl="https://www.medicinenet.com/common_medical_abbreviations_and_terms/article.htm"

response=requests.get(wikiurl)
print(response.status_code)

soup = BeautifulSoup(response.text, 'html.parser')
lists = soup.find_all('ul', {'class': ''})
lists
```

```python
abbrs = []
name = []
```

```python
for list in lists:
    for li in list.find_all('li'):
        # b tag
        b = li.find('b')
        if b:
            abbrs.append(b.text.replace(':', ''))

        # a tag
        a = li.find('a')
        if a:
            print(a.text)
            name.append(a.text)
```

```python
print(abbrs)
print(name)
```

```python
df = pd.DataFrame({'abbr': abbrs})
```

```python
df.to_csv('../dataset/csv/medical_abbreviations_medicinenet.csv', index=False)
```
