```python
import pandas as pd
import numpy as np
import requests # library to handle requests
from bs4 import BeautifulSoup # library to parse HTML documents
```

```python
wikiurl="https://www.asha.org/practice-portal/professional-issues/documentation-in-health-care/common-medical-abbreviations/"
table_class="table table-striped"
response=requests.get(wikiurl)
print(response.status_code)

soup = BeautifulSoup(response.text, 'html.parser')
table = soup.find('table',{'class':"table table-striped"})
```

```python
table
```

```python
df=pd.read_html(str(table))

# convert list to dataframe
df=pd.DataFrame(df[0])

df.head()
```

```python
df.to_csv('../dataset/csv/medical_abbreviations_asha.csv', index=False)
```

```python
df_medline = pd.read_csv('../dataset/csv/medical_abbreviations_asha.csv')
```
