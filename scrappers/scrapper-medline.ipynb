```python
wikiurl="https://medlineplus.gov/appendixb.html"
table_class="table-box"
response=requests.get(wikiurl)
print(response.status_code)

soup = BeautifulSoup(response.text, 'html.parser')
table = soup.find('table',{'class':"table-box"})
```

```python
table
```

```python
df=pd.read_html(str(table))

# convert list to dataframe
df=pd.DataFrame(df[0])

df.head()
```

```python
df.to_csv('medical_abbreviations_medlineplus.csv', index=False)
```
